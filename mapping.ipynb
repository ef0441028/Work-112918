```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
from astropy.utils.data import get_pkg_data_filename
import os
import shutil
from astropy.table import QTable
import matplotlib.colors
from matplotlib.mlab import bivariate_normal
```

```python
t_all = QTable.read('throughput_merged_data_all.csv', format='ascii.csv')
t_21 = QTable.read('throughput_merged_data_21.csv', format='ascii.csv')
t_22 = QTable.read('throughput_merged_data_22.csv', format='ascii.csv')
t_23 = QTable.read('throughput_merged_data_23.csv', format='ascii.csv')
t_24 = QTable.read('throughput_merged_data_24.csv', format='ascii.csv')
t_25 = QTable.read('throughput_merged_data_25.csv', format='ascii.csv')
```

```python
x = t_all['X']
y = t_all['Y']
z = t_all['Throughput']
fig,ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
plt.xlabel('Fiber Position X')
plt.ylabel('Fiber Position Y')
plt.title('Throughput at Focal Plane for all Fiber Harnesses')

colormap = plt.cm.bwr #or any other colormap
normalize = matplotlib.colors.Normalize(vmin=-1, vmax=1)
plt.scatter(x, y, c=z, s=15, cmap=colormap, norm=normalize);
fig.savefig('plotall.jpg')
```
